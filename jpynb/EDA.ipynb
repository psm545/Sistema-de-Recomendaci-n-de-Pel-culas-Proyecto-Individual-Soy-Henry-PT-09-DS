## **MAPA DE CALOR**

```python

# Seleccionar solo las columnas relevantes para el análisis
columns_of_interest = ['budget', 'popularity', 'revenue', 'runtime', 'vote_average', 'vote_count', 'return', 'release_year', 'release_month']
data_relevant = data[columns_of_interest]

# Calcular la matriz de correlación
correlation_matrix = data_relevant.corr()

# Crear un mapa de calor
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Mapa de Calor de Correlaciones')
plt.show()
```

```python

# Función para contar elementos en listas
def count_list_elements(series):
    counter = Counter()
    for item in series.dropna():
        # Convertir string de lista a lista real si es necesario
        if isinstance(item, str):
            item = ast.literal_eval(item)
        counter.update(item)
    return counter

""" 
Propósito:

Contar elementos en listas: Esta función cuenta la frecuencia de cada elemento en una columna de listas (por ejemplo, genres_names y company_names).
Conversión y conteo: Si los elementos están en formato de cadena que representa una lista, se convierten en listas reales y luego se cuenta la frecuencia de cada elemento usando Counter del módulo collections.

"""

# 3. Top 10 géneros más comunes
plt.figure(figsize=(12, 6))
genre_counts = count_list_elements(data['genres_names'])
top_genres = dict(sorted(genre_counts.items(), key=lambda x: x[1], reverse=True)[:10])

plt.barh(list(top_genres.keys()), list(top_genres.values()))
plt.title('Top 10 géneros más comunes')
plt.xlabel('Número de películas')
plt.gca().invert_yaxis()  # Para que el género más común esté arriba
plt.tight_layout()
plt.show()

# 4. Top 10 compañías productoras más prolíficas
plt.figure(figsize=(12, 6))
company_counts = count_list_elements(data['company_names'])
top_companies = dict(sorted(company_counts.items(), key=lambda x: x[1], reverse=True)[:10])

plt.barh(list(top_companies.keys()), list(top_companies.values()))
plt.title('Top 10 compañías productoras más prolíficas')
plt.xlabel('Número de películas')
plt.gca().invert_yaxis()  # Para que la compañía más prolífica esté arriba
plt.tight_layout()
plt.show()
```

```python

# Función para extraer géneros y crear una matriz de géneros
def extract_genres(row):
    genres = ast.literal_eval(row) if isinstance(row, str) else row
    return pd.Series({genre: 1 for genre in genres})
    
""" 
Propósito:

Extraer Géneros: La función extract_genres toma una fila que contiene una lista de géneros (en formato de cadena o lista) y devuelve una Serie donde cada género se convierte en una columna con valor 1. Esto facilita la creación de una matriz de géneros.
"""
# Crear una matriz de géneros
genre_matrix = data['genres_names'].apply(extract_genres).fillna(0)

"""
Matriz de Géneros: Aplica la función extract_genres a la columna genres_names para crear una matriz binaria (0 o 1) donde las columnas representan géneros y las filas representan películas. Las celdas contienen 1 si el género está presente para esa película, de lo contrario 0.
"""

# Calcular el número de películas y la calificación promedio por género
genre_stats = pd.DataFrame({
    'count': genre_matrix.sum(),
    'avg_rating': (genre_matrix * data['vote_average'].values[:, None]).sum() / genre_matrix.sum()
})

"""
Propósito:

Estadísticas por Género: Calcula dos métricas por género:
Número de Películas (count): Suma de las columnas en la matriz de géneros, representando el número total de películas para cada género.
Calificación Promedio (avg_rating): Calcula la calificación promedio por género, ponderada por el número de películas de cada género.
"""
# Ordenar los géneros por número de películas
genre_stats = genre_stats.sort_values('count', ascending=False)


# Seleccionar los top 15 géneros más comunes
top_15_genres = genre_stats.head(15).index

# Crear una matriz de correlación entre los top 15 géneros
genre_correlation = genre_matrix[top_15_genres].corr()

"""
Propósito:

Matriz de Correlación: Calcula la correlación entre los géneros seleccionados. Esto te ayuda a entender cómo se relacionan entre sí los géneros más comunes.
"""

# Crear el mapa de calor
plt.figure(figsize=(12, 10))
sns.heatmap(genre_correlation, annot=True, cmap='YlOrRd', vmin=0, vmax=1, center=0.5)
plt.title('Mapa de calor de correlación entre géneros de películas')
plt.tight_layout()
plt.show()

# Crear un mapa de calor que muestre la relación entre género, número de películas y calificación promedio
plt.figure(figsize=(12, 10))
genre_stats_normalized = (genre_stats - genre_stats.min()) / (genre_stats.max() - genre_stats.min())
sns.heatmap(genre_stats_normalized.loc[top_15_genres].T, annot=genre_stats.loc[top_15_genres].T.round(2), 
            fmt='g', cmap='YlOrRd', cbar_kws={'label': 'Valor normalizado'})
plt.title('Mapa de calor de géneros de películas: Cantidad vs Calificación promedio')
plt.tight_layout()
plt.show()
```
